```python
import re
import nltk
```

#### From Jurafsky & Martin, Speech and Language Processing (2021), Chapter 2
Write regular expressions for the following languages.
1. the set of all alphabetic strings;
2. the set of all lower case alphabetic strings ending in a b;
3. the set of all strings from the alphabet a,b such that each a is immediately preceded by and immediately followed by a b;

```python
wordlist = nltk.corpus.words.words('en')
wordlist
```

```python
list = ['123', 'word', 'hat']
```

```python
# the set of all alphabetic strings
alph_strings = [w for w in list if re.search(r'[\w]+', w)]
alph_strings
```

```python
# the set of all lower case alphabetic strings ending in a b
alph_strings_lower = [w for w in wordlist if re.search(r'^[a-z]*b$', w)]
alph_strings_lower
```

```python
# the set of all strings from the alphabet a,b such that each 'a' is immediately preceded 
# by and immediately followed by a 'b'
a_b = [w for w in wordlist if re.search(r'[^Aa][Bb][Aa][Bb][^Aa]', w)]
# if there is an a, it must be preceded by b
a_b
```
