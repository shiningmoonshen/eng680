# Regular Expressions for Detecting Word Patterns

Regular expressions provide tools for finding patterns in text. 
In order to use regular expressions, you need to import the relevant library (import re). We'll be using a list of words from nltk to perform search.

```python
import nltk
nltk.download('words')
```

```python
nltk.corpus.words.words('en')
```

```python
import re, nltk
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
```

## Using Basic Metacharacters 

 - re.search(p, s) checks whether the pattern p is in the string s.
 - the dollar sign ($) in the context of a regular expression matches the end of the word.
 - the caret symbol ^ matches the start of a string.
 - the wildcard symbol . matches any character.
 - the ? symbol specifies that the previous character is optional.

```python
re.search('ed$', 'education')
wordlist = [w for w in nltk.corpus.words.words('en')]
```

```python
#Finding words that end with -ed
my_ed = [w for w in wordlist if re.search('ed$', w)]
my_ed[:10]
```

```python
#Using the wildcard symbol to find words that have j as their 3rd letter, and t as their 6th letter.
my_jete = [w for w in wordlist if re.search('^..j..t..$', w)]
my_jete[:10]
```

Textonyms: words that have the same sequence of keystrokes. How can we find textonyms using regular expressions?

The ^ symbol matches the beginning of the word, the $ symbol matches the end of the word.

```python
#Regular expressions to find textonyms with 4, 6, 5, 3.
[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
```

```python
#To find words that only use 4, 5, and 6 as keys.
[w for w in wordlist if re.search('^[g-o]+$', w)][:10]
```

```python
#To find words that only use 2, 3, 5, and 6 as keys.
[w for w in wordlist if re.search('^[a-fj-o]+$', w)][:10]
```

The '+' and '' symbols are referred to as Kleene closures. 

The '+' symbol used in the previous code means "one or more instances of the preceeding item."

```python
nltk.download('nps_chat')
```

```python
#The + symbol can be applied to individual letters
chat_words = sorted(set(w for w in nltk.corpus.nps_chat.words()))
[w for w in chat_words if re.search('^m+i+n+e+$', w)]
```

```python
#The + symbol can also be applied to sets of letters enclosed in square brackets.
[w for w in chat_words if re.search('^[ha]+$', w)]
```

```python
[w for w in chat_words if re.search('^h+a+$', w)]
```

```python
[w for w in chat_words if re.search('^[ha]+$',w)]
```

The * symbol means "zero or more instances of the preceding item."

```python
# '^m*i*n*e*$' will match all words returned by '^m+i+n+e+$' but also words that lack 
# certain characters, like 'me'
my_mine = [w for w in chat_words if re.search('^m*i*n*e*$', w)]
my_mine[:10]
```

When the ^ sign appears inside square brackets it means that the characters in the brackets are excluded from search. Thus, '[^aeiouAEIOU]' would excludes vowels from search and return strings composed of non-vowel characters only.

```python
non_vowel = [w for w in chat_words if re.search('^[^aeiouAEIOU]+$', w)]
non_vowel[30:50]
```

```python
nltk.download('treebank')
```

```python
#Create a list of words to be searched with regular expressions
wsj = sorted(set(nltk.corpus.treebank.words()))
```

```python
# The escape symbol \ means that the following character does not have its special meaning
# within the regular expression. \ . matches a period.
my_numbers = [w for w in wsj if re.search(r'^[0-9]+\.[0-9]+$', w)]
my_numbers[:10]
```

```python
#Returns alphabetic characters followed by a $ sign
my_dollars = [w for w in wsj if re.search(r'^[A-Z]+\$$', w)]
my_dollars[:10]
```

```python
#Returns exactly 4 numeric characters
my_years = [w for w in wsj if re.search('^[0-9]{4}$', w)]
my_years[:10]
```

```python
#Returns a compound of the form number-word, when the number and word have between 3 and 5 characters
my_number_compounds = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
my_number_compounds
```

```python
#Returns compounds of the form word1-word2-word3, where word1 is 5 or more characters long
#word2 is between 2 and 3 characters long, and word3 is no more than 6 characters long. 
compounds = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
compounds[:10]
```

```python
#Returns sequencing ending in either -ed or -ing.
ed_ing = [w for w in wsj if re.search('(ed|ing)$', w)]
ed_ing[15:25]
```

```python
#If you omit square brackets around ed|ing$, the search returns words that 
#contain -ed in any position, not only word-final (cf. Biedermann)
ed_ing_2 = [w for w in wsj if re.search('ed|ing$', w)]
ed_ing_2[15:25]
```

## More examples

This examples below are from 'Automate the Bording Stuff with Python' by A. Sweigart. Chapter 7.

```python
# Create variable phoneNumRegEx; re.copile creates a Regex object.
phoneNumRegEx = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d')
```

```python
# Variable mo stores the result of search, which is called on phoneNumRegex.
mo = phoneNumRegEx.search('My number is 415-555-4242.')
# group() is called on mo and returns the match.
print('Phone number found: '+ mo.group())
```
